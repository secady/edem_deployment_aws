```python
sup_list = soup.select("div.elementor-widget-wrap section.projects-grid article") # Lista de Start-ups
```

```python
len(sup_list)
```

```python
sup_list[720].select("div.projects-grid__topic")[0].get_text()
```

```python
topic_list = []
for startup in sup_list:
    if startup.select("div.projects-grid__topic") == []:
        topic_list.append(None)
    else:
        pre_topic_list = [topic.get_text() for topic in startup.select("div.projects-grid__topic")]
        topic_list.append(pre_topic_list)
```

```python
topic_list
```

```python
sup_list[0].select("figure a img")[0].get("alt")
```

```python
link_list = [startup.find("a").get("href") for startup in sup_list]
```

```python
link_list
```

```python
logo_link_list = [startup.select("figure a img")[0].get("data-lazy-src") for startup in sup_list]
```

```python
logo_link_list
```

```python
name_list = [startup.select("figure a img")[0].get("alt") for startup in sup_list]
```

```python
name_list
```

```python
" ".join(sup_list[0].select("div.projects-grid__description")[0].get_text().split())
```

```python
text_list = [" ".join(startup.select("div.projects-grid__description")[0].get_text().split()) for startup in sup_list]
```

```python
text_list
```

```python
sup_list[720].select("div.projects-grid__phase")
```

```python
phase_list = []
for startup in sup_list:
    phase = startup.select("div.projects-grid__phase")
    if phase == []:
        phase_list.append(None)
    else:
        phase_list.append(phase[0].get_text())
```

```python
phase_list
```

```python
df = pd.DataFrame({
    "name": name_list,
    "phase": phase_list,
    "topics": topic_list,
    "description": text_list,
    "url": link_list,
    "logo_link": logo_link_list
})
```

```python
df
```

```python
df.to_json("startups_data.json", index=True)
df.to_json("startups_by_rows.json", orient= "index")
```
