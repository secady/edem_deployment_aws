```python
new_col = list(range(0,df.shape[0]))
```

```python
id_col = pd.Series(new_col, name="id_key", index=None)
```

```python
df2 = pd.concat([id_col, df], axis=1)
```

```python

meta = db.MetaData()
scrapped = db.Table(
    'scrapped',
    meta,
    db.Column('id_key', db.Integer, primary_key=True),
    db.Column('name', db.String),
    db.Column('phase', db.String),
    db.Column('topics', db.String),
    db.Column('description', db.String),
    db.Column('url', db.String),
    db.Column('logo_link', db.String)
    )
```

```python
meta.create_all(engine)
```

```python
data_insert = df2.to_dict(orient="records")
```

```python
data_insert
```

```python
with engine.connect() as connection:  
    connection.execute(scrapped.delete())
    connection.commit()
    connection.close()
```

```python
with engine.connect() as connection:  
    connection.execute(scrapped.insert(), data_insert)
    connection.commit()
    connection.close()
```
