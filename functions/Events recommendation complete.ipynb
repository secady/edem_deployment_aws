## Events recommendation class

Creates a list of all the events available ordenated by the most to the least interesting for the student. To do so, it performs a puctuation system where, for each interest in common between the student and the event, sums 0.333 to the puctuation, it calculates this for all events for all students, creating a dataframe. Finally returns an ordenated list of events from higher to lower for each student.

```python
import pandas as pd
import numpy as np

class EventRecommendation:
    def __init__(self, student_df, events_df):
        self.student_df = student_df
        self.events_df = events_df
        self.event_recommendation_df = None

    def drop_columns(self):
        self.student_df = self.student_df.drop(['Unnamed: 0', 'gender', 'age', 'year','category_id','name','surname','email','programme'], axis=1).copy()
        self.events_df = self.events_df.drop(['title','description','date'], axis=1).copy()
    def calculate_punctuation(self):
        self.event_recommendation_df = pd.DataFrame(columns=['student_id'] + list(self.events_df['id']))

        for student in self.student_df.itertuples():
            student_id = student.student_id
            punctuation = []

            for event in self.events_df.itertuples():
                event_name = event.id
                common_interests = sum([student[index] and event[index] for index in range(3, len(student))])
                punctuation.append(common_interests * 0.333)

            self.event_recommendation_df.loc[len(self.event_recommendation_df)] = [student_id] + punctuation

    def get_sorted_events_list(self):
        sorted_events_list = []
        for student in self.event_recommendation_df.itertuples():
            student_id = student.student_id
            events_sorted_indices = np.argsort(student[2:])
            events_sorted = [self.event_recommendation_df.columns[index+1] for index in events_sorted_indices]
            sorted_events_list.append((student_id, events_sorted[::-1]))

        return sorted_events_list


# Create instances of the class and use its methods
student_df = pd.read_csv('students_complete_df.csv')
events_df = pd.read_csv('events_complete_df.csv')

event_rec = EventRecommendation(student_df, events_df)
event_rec.drop_columns()
event_rec.calculate_punctuation()
sorted_events = event_rec.get_sorted_events_list()

# Print the sorted events for each student
for student_events in sorted_events:
    student_id, events_sorted = student_events
    print(f"Student id: {student_id}: Events recommended in order {events_sorted}")
```
